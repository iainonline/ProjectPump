```python
# example of a dataframe format and sample data of skater on a half pipe

import pandas as pd

#create a dataframe of x,y,z coordinates for two acceloerometers sensors with a time seperator value the time seperator value is 0.2 seconds
#the range of acceleration values is -50 to 50 m/s^2
#the patern of acceleration is a sine wave over 10 seconds for each cycle
#it replcate the motion of a skater on a half pipe

#create the columns for the dataframe

dfTwoSensors = pd.DataFrame(columns=['time','x', 'y', 'z', 'a','b','c'])

#set the time seperator value
timeSeperator = 0.2

#set the range of acceleration values
accelerationRange = 50

#set the number of cycles
numberOfCycles = 25

#set the sample data size
sampleDataSize = 10
```

```python
#create the data for the dataframe for number of cycles programatically

import math


for i in range(numberOfCycles):
    for j in range(sampleDataSize):
        # create the data for the dataframe for each cycle programatically
        # the x,y,z values are a sine wave with range of acceleration values accelerationRange
        # the a,b,c values are the same as x,y,z sine wave offset by 2 seconds
        # the range of acceleration values is accelerationRange
        # the time seperator value is timeSeperator
        # the number of cycles is numberOfCycles
        # the sample data size is sampleDataSize
        # use concatenation to add the data to the dataframe
        # use math.sin() to calculate the sine wave and a variable amp to determine amplitude
        
        dfTwoSensors = pd.concat([dfTwoSensors, pd.DataFrame({"time": float(i*sampleDataSize*timeSeperator + j*timeSeperator), "x": float(accelerationRange*math.sin(i*sampleDataSize*timeSeperator + j*timeSeperator)), "y": float(accelerationRange*math.sin(i*sampleDataSize*timeSeperator + j*timeSeperator)), "z": float(accelerationRange*math.sin(i*sampleDataSize*timeSeperator + j*timeSeperator)), "a": float(accelerationRange*math.sin(i*sampleDataSize*timeSeperator + j*timeSeperator + 2)), "b": float(accelerationRange*math.sin(i*sampleDataSize*timeSeperator + j*timeSeperator + 2)), "c": float(accelerationRange*math.sin(i*sampleDataSize*timeSeperator + j*timeSeperator + 2))}, index=[0])])
```

```python
#display the dataframe
print(dfTwoSensors)
```

```python
#plot the dataframe

import plotly.graph_objects as go
# zoom to a level where the changes in acceleration can be seen
# plot x,y,z axis labels as a separate series called board
# plot a,b,c axis labels as a separate series called skater
# include time axis label
# include a title for the plot 'Plot of sample accelerometer data from the board and the sensor on the skater'
# include a legend
# plot the dataframe using plotly
fig = go.Figure()
fig.update_layout(
    title="Plot of sample accelerometer data from the board and the sensor on the skater",
    xaxis_title="time",
    yaxis_title="acceleration",
    legend_title="Legend",
    font=dict(
        family="Courier New, monospace",
        size=18,
        color="RebeccaPurple"
    )
)
fig.add_trace(go.Scatter(x=dfTwoSensors['time'], y=dfTwoSensors['x'], name="board x"))
fig.add_trace(go.Scatter(x=dfTwoSensors['time'], y=dfTwoSensors['y'], name="board y"))
fig.add_trace(go.Scatter(x=dfTwoSensors['time'], y=dfTwoSensors['z'], name="board z"))
fig.add_trace(go.Scatter(x=dfTwoSensors['time'], y=dfTwoSensors['a'], name="skater x"))
fig.add_trace(go.Scatter(x=dfTwoSensors['time'], y=dfTwoSensors['b'], name="skater y"))
fig.add_trace(go.Scatter(x=dfTwoSensors['time'], y=dfTwoSensors['c'], name="skater z"))
fig.show()
```
